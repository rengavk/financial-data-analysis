Financial Data Analysis

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:/Users/DELL/Downloads/Finance_data.csv")
```

```python
for col in data.columns:
    print(col)
```

```python
data.head()
```

```python
summary_stats = data.describe()
summary_stats
```

```python
data.dropna(inplace=True)
```

```python
investment_columns = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 'Stock_Marktet']
investment_distribution = data[investment_columns].sum()
investment_distribution
```

```python
savings_objectives_count = data['What are your savings objectives?'].value_counts()
savings_objectives_count
```

```python
data.drop(['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'PPF', 'Fixed_Deposits',"Gold" ], axis = 1, inplace = True)
```

```python
data = data.replace(r'^\s*$', np.nan, regex = True)
```

```python
sns.countplot(x='age', data = data, palette = "Set3", linewidth = 2, edgecolor = "black", legend=False)
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x=data['gender'], hue = data['Investment_Avenues'], palette = 'summer', linewidth = 3, edgecolor = "white")
plt.title("INVESTMENT AVENUE")

plt.subplot(1,2,2)
sns.countplot(x= data['gender'], hue = data['Stock_Marktet'], palette = 'hot', linewidth = 3, edgecolor = 'white')
plt.title("STOCK MARKET")
plt.show()
```

```python
sns.histplot(data['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x=data['Factor'], palette = 'coolwarm', linewidth = 2, edgecolor = "black")
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.pointplot(x = "Purpose", y = "age", data = data, linestyles = "--", capsize = .3, color = "Blue")
plt.show()
```

```python
plt.figure(figsize= (18,6))
plt.subplot(1,2,1)
sns.countplot(hue = data['gender'], x = data['Duration'], palette = 'viridis', linewidth = 2, edgecolor = "black" )
plt.title("DURATION")

plt.subplot(1,2,2)
sns.countplot(hue = data["gender"], x = data["Invest_Monitor"], palette = "seismic", linewidth = 2, edgecolor ="black")
plt.title("INVESTMENT MONITORING")
plt.show()
```

From the Graph, it's clear that Male invest for longer duration than Female whereas both of them invest mostly monthly than weekly or daily basis.
It's very clear that Male around 26-29 are investing a lot and in general people are considering high returns, low risk and best savings for future as a main factors. 
Stock Market is being the most chosen method of investing by large number of male group
